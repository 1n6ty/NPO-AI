```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

"""Файл 1"""
d_set_clnts = pd.read_csv("data/train_dataset_train_data_npo/npo_clnts.csv")

"""Файл 2"""
d_set_cntrbtrs = pd.read_csv("data/train_dataset_train_data_npo/npo_cntrbtrs.csv")

"""Файл 3"""
d_set_trnsctns = pd.read_csv("data/train_dataset_train_data_npo/npo_trnsctns.csv")
```

```python
"""Пол клиента: "м": "0", "ж": "1" """
sns.histplot(data=d_set_clnts, x="gndr")
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), p.get_height(), ha='center', va='bottom')
```

```python
"""Живет клиент в городе (1) или нет (0)"""
sns.histplot(data=d_set_clnts, x="city")
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), p.get_height(), ha='center', va='bottom')
```

```python
"""Возраст"""
plt.figure(figsize = (20, 6))
sns.histplot(data=d_set_clnts, x="age")
ax = plt.gca()
ax.set(xlim=(10, 125))
```

```python
"""Статус счета клиента: "Закрыт": "0", "Открыт": "1" """
sns.histplot(data=d_set_cntrbtrs, x="npo_accnt_status")
ax = plt.gca()
ax.set()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), p.get_height(), ha='center', va='bottom')
```

```python
"""Число взносов"""
plt.figure(figsize = (20, 6))
sns.histplot(data=d_set_cntrbtrs, x="npo_pmnts_nmbr")
ax = plt.gca()
ax.set()
```

```python
"Балланс счёта"
blnc = d_set_cntrbtrs["npo_blnc"]

plt.figure(figsize = (30, 6))

# Создание промежутков с помощью функции pd.cut()
bins = pd.cut(blnc, bins=[0, 100, 500, 1000, 2500, 5000, 10000, 20000, 40000, 60000, 100000, 200000, 500000, 1e6])

# Добавление столбца с промежутками в таблицу
blnc['Range'] = bins

# Создание графика с помощью seaborn
sns.countplot(x='Range', data=blnc)

# Добавление числовых значений над колонками
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), p.get_height(), ha='center', va='bottom')
```

```python
"""Сумма взносов"""
sum = d_set_cntrbtrs["npo_pmnts_sum"]

plt.figure(figsize = (30, 6))

# Создание промежутков с помощью функции pd.cut()
bins = pd.cut(sum, bins=[0, 100, 500, 1000, 2500, 5000, 10000, 20000, 40000, 60000, 100000, 200000, 500000, 1e7])

# Добавление столбца с промежутками в таблицу
sum['Range'] = bins

# Создание графика с помощью seaborn
sns.countplot(x='Range', data=sum)

# Добавление числовых значений над колонками
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), p.get_height(), ha='center', va='bottom')
```

```python
"""Сумма дохода"""
incm = d_set_cntrbtrs["npo_ttl_incm"]

plt.figure(figsize = (30, 6))

# Создание промежутков с помощью функции pd.cut()
bins = pd.cut(incm, bins=[0, 100, 500, 1000, 2500, 5000, 10000, 20000, 40000, 60000, 100000, 200000, 500000, 1e7])

# Добавление столбца с промежутками в таблицу
incm['Range'] = bins

# Создание графика с помощью seaborn
sns.countplot(x='Range', data=incm)

# Добавление числовых значений над колонками
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), p.get_height(), ha='center', va='bottom')
```

```python
"""Тип операции по счету клиента: "Поступление взносов": "0", "Начисление дохода": "1" """
sns.histplot(data=d_set_trnsctns, x="npo_operation_group")
ax = plt.gca()
ax.set()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), p.get_height(), ha='center', va='bottom')
```

```python
"""Размер операции"""
sum = d_set_trnsctns["npo_sum"]

plt.figure(figsize = (30, 6))

# Создание промежутков с помощью функции pd.cut()
bins = pd.cut(sum, bins=[0, 100, 250, 500, 1000, 2500, 5000, 10000, 20000, 40000, 60000, 100000, 200000, 1e7])

# Добавление столбца с промежутками в таблицу
sum['Range'] = bins

# Создание графика с помощью seaborn
sns.countplot(x='Range', data=sum)

# Добавление числовых значений над колонками
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height(), p.get_height(), ha='center', va='bottom')
```
